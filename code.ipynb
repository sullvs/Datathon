```python
import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_csv('/Users/shahadaleissa/Datasets/equipment_failure_data_1.csv')

# Rename the 'ID' column to 'MACHINE_ID'
data = data.rename(columns={'ID': 'MACHINE_ID'})

# Add a ID column that is a unique identifier for each row
data['ID'] = range(1, len(data) + 1)

#bring the ID column to the front
cols = data.columns.tolist()
cols = cols[-1:] + cols[:-1]
data = data[cols]

data.head()
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import SMOTENC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
# Define features and target variable
X = data.drop(columns=['EQUIPMENT_FAILURE'])
y = data['EQUIPMENT_FAILURE']

# X['DATE'] = pd.to_datetime(X['DATE'], format='%m/%d/%y')
# X['YEAR'] = X['DATE'].dt.year
# X['MONTH'] = X['DATE'].dt.month
# X['DAY'] = X['DATE'].dt.day
# X = X.drop(columns=['DATE'])

# #convert REGION_CLUSTER and MAINTENANCE_VENDOR and MANUFACTURER to float 
# X['REGION_CLUSTER'] = X['REGION_CLUSTER'].astype('category').cat.codes
# X['MAINTENANCE_VENDOR'] = X['MAINTENANCE_VENDOR'].astype('category').cat.codes
# X['MANUFACTURER'] = X['MANUFACTURER'].astype('category').cat.codes


# Split the data into training and testing and validation sets (70% train, 20% test, 10% validation)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_test, X_eval, y_test, y_eval = train_test_split(X_test, y_test, test_size=0.1, random_state=42)

#combine the training data
train_data = pd.concat([X_train, y_train], axis=1)
#combine the testing data
test_data = pd.concat([X_test, y_test], axis=1)
#combine the evaluation data
eval_data = pd.concat([X_eval, y_eval], axis=1)
```

```python



# Handle class imbalance using SMOTE
smote = SMOTENC(random_state=12,  categorical_features=[0,1,2,3,4])
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

# Build a Random Forest Classifier model
model = RandomForestClassifier(random_state=42)
model.fit(X_train_res, y_train_res)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
